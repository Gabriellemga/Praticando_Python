# Desafio 1 - Analisando vendas em uma rede de lojas

Nesta atividade, você está trabalhando com dados de uma rede de lojas de departamentos e precisa explorar as características demográficas dos clientes para entender melhor seus hábitos de compra. Uma das perguntas chave para a análise demográfica é identificar a idade média dos clientes de acordo com o método de pagamento utilizado.

Os dados estão disponíveis na seguinte URL:

url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/loja_vendas.csv'

Sua tarefa é carregar os dados de vendas, explorar as informações disponíveis, e aplicar o método pivot_table() para calcular a média de idade dos clientes por método de pagamento.

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/loja_vendas.csv'
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df_pagamento_por_idade = pd.pivot_table(df, values = 'idade', index = 'metodo_pagamento' ).round()
```

```python
df_pagamento_por_idade.rename(columns = {'idade': 'Idade(média)'}, inplace = True)
df_pagamento_por_idade
```

# Desafio 4 - Reformatando a tabela de vendas por unidade e categoria de produto

Depois de calcular a soma do valor total de vendas por combinação de unidade e categoria de produto, a próxima etapa é reformatar essa tabela para facilitar comparações entre unidades.

Transforme a tabela criada anteriormente para que as categorias de produto sejam as colunas, e cada unidade represente uma linha. Isso facilitará a visualização do desempenho de vendas de cada categoria por unidade.

```python
df_soma = pd.pivot_table(df, index = 'unidade' , columns='categoria_produto', values = ['valor_total'], aggfunc = 'sum')
df_soma = df_soma.rename(columns = {'valor_total':'Soma total'})
df_soma
```

# Desafio 5 - Adicionando valores totais na tabela de vendas por unidade e categoria de produto

Seu desafio é incluir informações adicionais na tabela para aprimorar a análise de desempenho. Você deve incluir uma coluna que sumarize o valor total de vendas para cada unidade e adicionar uma linha ao final da tabela que mostre o total de vendas para cada categoria

Dica: Lembre-se que existe um parâmetro do método pivot_table() que pode ajudar nisso!

Essas adições ajudarão a visualizar não apenas o desempenho individual de cada unidade, mas também a contribuição de cada categoria para o total de vendas, oferecendo uma visão compreensiva do negócio.

```python
df_soma = pd.pivot_table(df, index = 'unidade' , columns='categoria_produto', values = ['valor_total'], aggfunc = 'sum', margins = True, margins_name = 'Total')
df_soma
```
